EXAMPLE 1 — Load and Analyze Iris Dataset

This code loads the Iris dataset and displays its basic information.

```python
# --------------------------------------------
# Example 1: Load and Analyze Iris Dataset
# --------------------------------------------

# Import required library
from sklearn.datasets import load_iris
import pandas as pd

# Load iris dataset
iris = load_iris()

# Convert to DataFrame for better visualization
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add target column
df["species"] = iris.target

# Print first 5 rows
print("First 5 rows of dataset:")
print(df.head())

# Print dataset information
print("\nDataset Shape:", df.shape)
print("\nFeature Names:", iris.feature_names)
print("\nTarget Names:", iris.target_names)
```

EXAMPLE 2 — Visualize Iris Dataset

This code plots a scatter graph between two features.

```python
# --------------------------------------------
# Example 2: Visualize Iris Dataset
# --------------------------------------------

# Import libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target

# Plot Sepal Length vs Sepal Width
plt.figure()

plt.scatter(X[:, 0], X[:, 1], c=y)

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Iris Dataset Visualization")

plt.show()
```

EXERCISE — Car and Cup Image Processing
This code:
Reads car & cup images
Resizes to 256x256
Converts to grayscale
Normalizes
Calculates mean, contrast, energy
Displays images

```python
# --------------------------------------------
# Exercise: Car and Cup Image Processing (Kaggle Fixed Version)
# --------------------------------------------

import cv2
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------------------------
# Step 1: Read images using FULL Kaggle path
# -------------------------------------------------

car = cv2.imread("/kaggle/input/datasets/abdulwasiaabdulwahid/lab4images/car.jpg")
cup = cv2.imread("/kaggle/input/datasets/abdulwasiaabdulwahid/lab4images/cup.jpg")

# Check if images loaded properly
if car is None:
    print("Car image not loaded! Check path.")
if cup is None:
    print("Cup image not loaded! Check path.")

# -------------------------------------------------
# Step 2: Resize images to 256x256
# -------------------------------------------------

car = cv2.resize(car, (256, 256))
cup = cv2.resize(cup, (256, 256))

# -------------------------------------------------
# Step 3: Convert to grayscale
# -------------------------------------------------

car_gray = cv2.cvtColor(car, cv2.COLOR_BGR2GRAY)
cup_gray = cv2.cvtColor(cup, cv2.COLOR_BGR2GRAY)

# -------------------------------------------------
# Step 4: Normalize images (0 to 1 range)
# -------------------------------------------------

car_norm = car_gray / 255.0
cup_norm = cup_gray / 255.0

# -------------------------------------------------
# Step 5: Calculate features
# -------------------------------------------------

car_mean = np.mean(car_gray)
car_contrast = np.std(car_gray)
car_energy = np.sum(np.square(car_gray))

cup_mean = np.mean(cup_gray)
cup_contrast = np.std(cup_gray)
cup_energy = np.sum(np.square(cup_gray))

# Print features
print("Car - Mean:", car_mean)
print("Car - Contrast:", car_contrast)
print("Car - Energy:", car_energy)

print("Cup - Mean:", cup_mean)
print("Cup - Contrast:", cup_contrast)
print("Cup - Energy:", cup_energy)

# -------------------------------------------------
# Step 6: Display images
# -------------------------------------------------

plt.figure(figsize=(10,6))

plt.subplot(2,3,1)
plt.imshow(cv2.cvtColor(car, cv2.COLOR_BGR2RGB))
plt.title("Car Original")
plt.axis("off")

plt.subplot(2,3,2)
plt.imshow(car_gray, cmap='gray')
plt.title("Car Gray")
plt.axis("off")

plt.subplot(2,3,3)
plt.imshow(car_norm, cmap='gray')
plt.title("Car Normalized")
plt.axis("off")

plt.subplot(2,3,4)
plt.imshow(cv2.cvtColor(cup, cv2.COLOR_BGR2RGB))
plt.title("Cup Original")
plt.axis("off")

plt.subplot(2,3,5)
plt.imshow(cup_gray, cmap='gray')
plt.title("Cup Gray")
plt.axis("off")

plt.subplot(2,3,6)
plt.imshow(cup_norm, cmap='gray')
plt.title("Cup Normalized")
plt.axis("off")

plt.tight_layout()
plt.show()
```
