```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters
mu = 0.5
sigma = 1

# Generate data
data = np.random.normal(mu, sigma, 1000)

# Plot histogram
plt.hist(data, bins=30, density=True, label="Histogram")

# Plot PDF
x = np.linspace(-4, 5, 1000)
y = (1 / (sigma * np.sqrt(2 * np.pi))) * \
    np.exp(-(x - mu)**2 / (2 * sigma**2))

plt.plot(x, y, label="Normal PDF")

# Labels
plt.xlabel("Values")
plt.ylabel("Density")
plt.title("Normal Distribution")
plt.legend()

plt.show()
```

```python
# Example three
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)

def randrange(n, vmin, vmax):
    return (vmax - vmin) * np.random.rand(n) + vmin

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

n = 100
for m, zlow, zhigh in [('o', -70, -25), ('^', -30, -9)]:
    xs = randrange(n, 23, 32)
    ys = randrange(n, 0, 100)
    zs = randrange(n, zlow, zhigh)
    ax.scatter(xs, ys,zs, marker=m)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
#This code makes a 3D scatter graph with random dots floating in space. 
#It draws two types of dots, circles and triangles, at different positions on the X, Y, and Z axes. 
#Labels are added to show which axis is which, and plt.
#show() displays the graph on the screen. 
#The dots positions are random,
#so the graph looks different each time you run it.
```

4. Ecerxise 
Generate a random array of 100x5. Determine the 
correlation among the columns and then plot its
heatmap and the color 
bar. Hint:
matplotlib.pyplot.imshow(Data, cmap='hot',interpolation='nearest') and np.corrcoef(x, y)

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a random array of size 100x5
data = np.random.rand(100, 5)
# 100 rows (samples), 5 columns (variables)

# Step 2: Calculate correlation among columns
# np.corrcoef calculates the correlation matrix
# We set rowvar=False because columns represent variables
corr_matrix = np.corrcoef(data, rowvar=False)

# Step 3: Plot the correlation matrix as a heatmap
plt.imshow(corr_matrix, cmap='hot', interpolation='nearest')
# cmap='hot' -> color map from black (low) to red/yellow (high)
# interpolation='nearest' -> no smoothing, show exact values

# Step 4: Add color bar to interpret colors
plt.colorbar()

# Step 5: Add labels and title for clarity
plt.xticks(range(5), ['Col1', 'Col2', 'Col3', 'Col4', 'Col5'])
plt.yticks(range(5), ['Col1', 'Col2', 'Col3', 'Col4', 'Col5'])
plt.title("Heatmap of Correlation Matrix")

# Step 6: Display the plot
plt.show()
```

Task 2: Statistical Analysis
Using NumPy, calculate:
 Average marks of each student
 Average marks of each subject
 Highest and lowest marks obtained overall

```python
import numpy as np

# ----------------------------
# Task 1: Data Generation
# ----------------------------

np.random.seed(42)

# 10 students  5 subjects
marks = np.random.randint(0, 101, size=(10, 5))

print("Students' Marks (10  5):\n")
print(marks)


# ----------------------------
# Task 2: Statistical Analysis
# ----------------------------

# Average marks of each student
avg_per_student = np.mean(marks, axis=1)

# Average marks of each subject
avg_per_subject = np.mean(marks, axis=0)

# Highest and lowest marks overall
highest_mark = np.max(marks)
lowest_mark = np.min(marks)

subjects = [
    "Artificial Intelligence",
    "Data Science",
    "Machine Learning",
    "Python Programming",
    "Mathematics"
]

print("\nAverage Marks of Each Student:")
for i, avg in enumerate(avg_per_student, start=1):
    print(f"Student {i}: {avg:.2f}")

print("\nAverage Marks of Each Subject:")
for sub, avg in zip(subjects, avg_per_subject):
    print(f"{sub}: {avg:.2f}")

print("\nHighest Mark Overall:", highest_mark)
print("Lowest Mark Overall:", lowest_mark)
```

Task 3: Copy vs View Demonstration
 Create a subset of the original array using slicing.
 Modify the subset and observe the effect on the original array.
 Clearly mention whether the operation created a copy or a view.

```python
import numpy as np

# Original marks array (from Task 1)
np.random.seed(42)
marks = np.random.randint(0, 101, size=(10, 5))

print("Original Marks:\n", marks)

# ----------------------------
# Create a subset using slicing
# ----------------------------
subset_view = marks[0:3, 0:2]   # First 3 students, first 2 subjects

print("\nSubset (Before Modification):\n", subset_view)

# ----------------------------
# Modify the subset
# ----------------------------
subset_view[:, :] = 0

print("\nSubset (After Modification):\n", subset_view)

print("\nOriginal Marks After Modifying Subset:\n", marks)
```

```python
import matplotlib.pyplot as plt

# 1. Calculate Correlation Matrix (between subjects)
# rowvar=False tells NumPy to correlate columns (subjects) instead of rows
corr_matrix = np.corrcoef(marks, rowvar=False)

# 2. Plot Correlation Heatmap
subjects = ['AI', 'Data Sci', 'ML', 'Python', 'Math']
plt.imshow(corr_matrix, cmap='coolwarm', vmin=-1, vmax=1)

# 3. Add Colorbar and Labels
plt.colorbar(label='Correlation Scale')
plt.xticks(range(5), subjects)
plt.yticks(range(5), subjects)
plt.title('Subject Correlation Analysis')

plt.show()
```
