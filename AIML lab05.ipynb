```python
# Example 1 mean (average of nubmers)
import numpy 
speed = [99,86,67,66,111,86,103,87,94,78,77,85,86]
x = numpy.mean(speed)
print(x)
```

```python
# Example 2 median ( the between number)
import numpy
speed = [99,86,67,66,11,86,103,87,94,78,77,85,86]
x = numpy.median(speed)
print(x)
```

```python
# Example 3 Mode(most occur number)
import statistics
speed = [99,86,67,66,11,86,103,87,94,78,77,85,86]
x = statistics.mode(speed)
print(x)
```

```python
# Example 4 Std(Standard Deviation)
import numpy
speed = [86,87,88,86,87,85,86]
x = numpy.std(speed)
print (x)
```

```python
# Example 5 var(Variance)
import numpy 
speed = [32,111,138,28,59,77,97]
x = numpy.var(speed)
print (x)
```

```python
# Example 6 Percentiles
import numpy
ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]
x = numpy.percentile(ages, 90)
print ("90: ",x)
y = numpy.percentile(ages, 78)
print ("78: ",y)
```

```python
# Example 7 Data Distribution
import numpy 
x = numpy.random.uniform (0.0, 5.0, 250)
print (x)
```

```python
# Example 8 Histrogram
import numpy 
import matplotlib.pyplot as plt

x = numpy.random.uniform(1.0,5.0, 250)
plt.hist(x,15)
plt.show()
```

```python
# Example 9  Normal Data Distribuiton
import numpy 
import matplotlib.pyplot as plt

x = numpy.random.normal(5.0,1.0,100000)
plt.hist(x,100)
plt.show()
```
