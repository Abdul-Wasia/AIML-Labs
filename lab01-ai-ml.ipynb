```python
#1. Define "names" and "height" lists that contain the name and height of your friends 
#respectively. Then find the total number of entries in the list, maximum height, and 
#minimun height, and minimum height with their names, and find the average height of your
#friends.
```

```python
#1. define the lists
names = ["Farooq","Naveed","Sagar","Junaid","Ali"]
heights = [165,180,172,158,190]
#Heights in cm

#2. Total number of entries 
total_entries = len(names)

#3. Maximum and Minimum heights
max_height = max(heights)
min_height = min(heights)

#4. Find the names associated with max/min heights
#we find the index(position) of the value and look it up in the name list

max_name = names[heights.index(max_height)]
min_name = names[heights.index(min_height)]

#5. Calculate the average Height 
average_height = sum(heights)/total_entries

#Output the resluts
print(f"Total entries: {total_entries}")
print(f"Tallest friend:{max_name}({max_height}cm)")
print(f"Shortest friend: {min_name}({min_height}cm)")
print(f"Average height:{average_height:.2f}cm")
```

```python
#4. Create a list of oeven nnumbers and odd numbers with the help of for loop 
#and append function. Then combine both lists.
```

```python
# Initialize empty lists
even_numbers = []
odd_numbers = []

# Loop through from 1-20 
for num in range(1, 21):
    if num % 2 == 0:
        # If divisible by two then it's even.
        even_numbers.append(num)
    else: 
        # Added the missing colon here
        # Fixed "odd_number" to "odd_numbers"
        odd_numbers.append(num)

# Combining both lists using + operator
combined_list = even_numbers + odd_numbers

# Display Results
print("Even Numbers:", even_numbers)
print("Odd Numbers:", odd_numbers)
print("Combined List:", combined_list)
```

```python
"""7. Weather data of Pakistan's different regions is given in figure belw. create a nested 
disctionary for this data. Also write a functin that takes the region and type of data 
required as arguments and return with required data value."""
```

```python
# Creating the nested Dictionary for given weather data of Pakistan's different regions
weather_stats = {
    "Azad Kashmir": {"temp_max_C": 29.6, "temp_min_C": 16.0, "Sunshine_hours": 2884, "rainy_days": 67, "Precipitation": 1168, "Humidity_percent": 61.0},
    "Balochistan": {"temp_max_C": 32.0, "temp_min_C": 18.0, "Sunshine_hours": 3176, "rainy_days": 17, "Precipitation": 193, "Humidity_percent": 41.0},
    "Islamabad": {"temp_max_C": 28.9, "temp_min_C": 15.8, "Sunshine_hours": 2957, "rainy_days": 74, "Precipitation": 1296, "Humidity_percent": 61.0},
    "Khyber Pakhtunkhwa": {"temp_max_C": 29.1, "temp_min_C": 15.3, "Sunshine_hours": 2847, "rainy_days": 53, "Precipitation": 730, "Humidity_percent": 57.0},
    "Punjab": {"temp_max_C": 31.1, "temp_min_C": 17.8, "Sunshine_hours": 2592, "rainy_days": 44, "Precipitation": 624, "Humidity_percent": 55.0},
    "Sindh": {"temp_max_C": 34.5, "temp_min_C": 20.3, "Sunshine_hours": 3103, "rainy_days": 14, "Precipitation": 212, "Humidity_percent": 53.0}
}

# Function to retrieve weather data for a specific region and type of data
def get_weather_data(region, data_type):
    # Check if the region exists in the dictionary
    if region in weather_stats:
        # Check if the data_type is valid
        if data_type in weather_stats[region]:
            return weather_stats[region][data_type]
        else:
            return f"Invalid data_type. Valid data_types are: 'temp_max_C', 'temp_min_C', 'Sunshine_hours', 'rainy_days', 'Precipitation', 'Humidity_percent'."
    else:
        return f"Invalid region. Valid regions are: {', '.join(weather_stats.keys())}"

# Example usage
region = "Punjab"
data_type = "Humidity_percent"
result = get_weather_data(region, data_type)
print(f"The {data_type} for {region} is: {result}%")
```
