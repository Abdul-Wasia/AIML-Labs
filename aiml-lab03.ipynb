EXERCISE 1 – Titanic Dataset

```python
# --------------------------------------------
# Exercise 1: Titanic Dataset Analysis
# --------------------------------------------

import pandas as pd

# Create dataset again
data = {
    "survived": [0, 1, 1, 0, 1, 0, 1, 0],
    "class": ["Third", "First", "Third", "Second", "First", "Third", "Second", "First"],
    "sex": ["male", "female", "female", "male", "female", "male", "female", "male"],
    "age": [22, 38, 26, 35, 28, 2, 19, 40],
    "fare": [7.25, 71.83, 7.92, 53.1, 11.13, 21.07, 30.5, 80.0]
}

df = pd.DataFrame(data)

# --------------------------------------------
# (a) Number of people survived and not survived
# --------------------------------------------

survival_count = df["survived"].value_counts()

print("Survival Count:")
print(survival_count)

survival_count.plot(kind="bar", title="Survival Count")


# --------------------------------------------
# (b) Survived by class and gender
# --------------------------------------------

group_data = df.groupby(["class", "sex"])["survived"].sum()

print("\nSurvival by Class and Gender:")
print(group_data)

group_data.plot(kind="bar", title="Survival by Class and Gender")


# --------------------------------------------
# (c) Mean values grouped by gender and class
# --------------------------------------------

mean_values = df.groupby(["sex", "class"]).mean(numeric_only=True)

print("\nMean Values Grouped by Gender and Class:")
print(mean_values)
```

EXERCISE 2 – Random DataFrame

```python
# --------------------------------------------
# Exercise 2: Random DataFrame Operations
# This code creates a random DataFrame and 
# performs different operations step by step
# --------------------------------------------

# Import required libraries
import pandas as pd
import numpy as np

# --------------------------------------------
# Create random DataFrame (20 rows, 4 columns)
# --------------------------------------------

df = pd.DataFrame(np.random.rand(20, 4))

print("Original DataFrame:")
print(df)

# --------------------------------------------
# (a) Display last 3 rows
# --------------------------------------------

print("\nLast 3 Rows:")
print(df.tail(3))

# --------------------------------------------
# (b) Rename columns as col0, col1, col2, col3
# --------------------------------------------

df.columns = ["col0", "col1", "col2", "col3"]

print("\nAfter Renaming Columns:")
print(df.head())

# --------------------------------------------
# (c) Print values greater than 0.5 in first column
# --------------------------------------------

print("\nValues in col0 greater than 0.5:")
print(df[df["col0"] > 0.5])

# --------------------------------------------
# (d) Sort ascending and descending by col0
# --------------------------------------------

print("\nSorted Ascending by col0:")
print(df.sort_values(by="col0"))

print("\nSorted Descending by col0:")
print(df.sort_values(by="col0", ascending=False))

# --------------------------------------------
# (e) Access element at 4th row and 3rd column
# (Index starts from 0, so row=3, column=2)
# --------------------------------------------

element = df.iloc[3, 2]
print("\nElement at 4th row and 3rd column:", element)

# --------------------------------------------
# (f) Add two new columns: new1 and new2
# --------------------------------------------

df["new1"] = np.random.rand(20)
df["new2"] = np.random.rand(20)

print("\nAfter Adding new1 and new2:")
print(df.head())

# --------------------------------------------
# (g) Delete first column (col0)
# --------------------------------------------

df.drop("col0", axis=1, inplace=True)

print("\nAfter Deleting col0:")
print(df.head())

# --------------------------------------------
# (h) Print number of rows and columns
# --------------------------------------------

print("\nShape of DataFrame (rows, columns):")
print(df.shape)

# --------------------------------------------
# (i) Print descriptive statistics
# --------------------------------------------

print("\nDescriptive Statistics:")
print(df.describe())

# --------------------------------------------
# (j) Add a column with two NaN values
# then drop rows containing NaN
# --------------------------------------------

# Add column with NaN values
df["nan_col"] = np.nan

# Insert only two NaN values (others will remain NaN by default)
df.loc[0, "nan_col"] = np.nan
df.loc[1, "nan_col"] = np.nan

print("\nBefore Dropping NaN:")
print(df.head())

# Drop rows with NaN values
df_clean = df.dropna()

print("\nAfter Dropping NaN:")
print(df_clean)
```

EXERCISE 3 – Titanic String Operations

```python
# --------------------------------------------
# Exercise 3: String Operations
# --------------------------------------------

import pandas as pd

# Create dataset with names
data = {
    "Name": ["Braund, Mr. Owen", 
             "Cumings, Mrs. John", 
             "Heikkinen, Miss. Laina",
             "Allen, Mr. William"]
}

df = pd.DataFrame(data)

# (a) Lowercase names and split by comma

df["Name"] = df["Name"].str.lower()
df["Name_split"] = df["Name"].str.split(",")

print(df)


# (b) Length of longest name

df["Name_length"] = df["Name"].str.len()

longest_name = df["Name_length"].max()

print("\nLength of longest name:", longest_name)
```
